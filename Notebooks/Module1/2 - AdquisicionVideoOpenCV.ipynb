#### 1. CARGA DE LIBRERIAS

```python
# ============================================
# LIBRERÍAS
# ============================================
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

%matplotlib inline
```

#### 2 — FUNCIÓN EFECTO CARTOON

```python
# ============================================
# FUNCIÓN: EFECTO CARTOON
# ============================================
def Cartoon(image_color):
    """
    Aplica un efecto tipo caricatura (stylization)
    
    Parámetros:
    image_color: imagen en formato BGR
    
    Retorna:
    Imagen estilizada
    """
    
    # sigma_s controla el tamaño de las regiones
    # sigma_r controla la intensidad del efecto
    dst = cv2.stylization(image_color, sigma_s=60, sigma_r=0.5)
    
    return dst
```

#### 3 — INICIALIZAR CÁMARA

```python
# ============================================
# INICIALIZACIÓN DE LA CÁMARA
# ============================================

# 0 = cámara principal
cap = cv2.VideoCapture(0)

# Validación
if not cap.isOpened():
    raise RuntimeError("No se pudo abrir la cámara")
```

#### 4 — LOOP DE VIDEO (TIEMPO REAL)

```python
# ============================================
# CAPTURA DE VIDEO EN TIEMPO REAL
# ============================================

while True:
    
    # Leer frame de la cámara
    ret, frame = cap.read()
    
    # Validar lectura
    if not ret:
        print("Error al leer frame")
        break
    
    # Aplicar efecto cartoon / seleccionar 
    cartoon_frame = Cartoon(frame)     

    # Mostrar video procesado
    cv2.imshow("Cartoon Effect", cartoon_frame)

    edges = cv2.Canny(cv2.GaussianBlur(
    cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY),
    (5,5), 0), 50, 150)

    cv2.imshow("Edges", edges)
    
    # Mostrar video original
    cv2.imshow("Webcam Video", frame)
    
    
    
    # Presionar ENTER para salir
    if cv2.waitKey(1) == 13:
        break
```

#### 5 — LIBERAR RECURSOS

```python
# ============================================
# LIBERAR RECURSOS
# ============================================

cap.release()
cv2.destroyAllWindows()
```

#### 6 — EXPLICACIÓN (MARKDOWN)

## 📌 Flujo del sistema

Webcam → VideoCapture → Frame → Procesamiento → Visualización

## 🔍 Explicación

- `VideoCapture(0)` → accede a la cámara
- `cap.read()` → obtiene cada frame
- `Cartoon()` → aplica efecto visual
- `imshow()` → muestra resultado
- `waitKey()` → controla salida
