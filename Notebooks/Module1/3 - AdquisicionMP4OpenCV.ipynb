#### 2 — CARGAR VIDEO MP4

got video from https://www.pexels.com/search/videos/dashcam%20road%20driving%E2%80%9D/

```python
# ============================================
# CARGA DEL ARCHIVO MP4
# ============================================

from pathlib import Path
import cv2

VIDEO_PATH = Path("../../data/Video/5382495-uhd_3840_2160_24fps.mp4")

print("Exists:", VIDEO_PATH.exists())
print("Absolute path:", VIDEO_PATH.resolve())

cap = cv2.VideoCapture(str(VIDEO_PATH))

if not cap.isOpened():
    raise RuntimeError("No se pudo abrir el archivo de video")

print("Video opened correctly")
```

#### 3 — LECTURA DE FRAMES

```python
# ============================================
# LECTURA FRAME POR FRAME
# ============================================

while cap.isOpened():
    
    # ret: indica si el frame fue leído correctamente
    # frame: imagen actual del video
    ret, frame = cap.read()
    
    # Si no hay más frames (fin del video)
    if not ret:
        print("Fin del video o error de lectura")
        break
    
    # Mostrar frame
    cv2.imshow("Test MP4", frame)
    
    # Presionar ENTER (13) para salir
    if cv2.waitKey(1) == 13:
        break
```

#### 4 — LIBERAR RECURSOS

```python
# ============================================
# LIBERAR RECURSOS
# ============================================

cap.release()
cv2.destroyAllWindows()
```

#### 5 — EXPLICACIÓN (MARKDOWN)

```python
## 📌 Flujo del sistema

Archivo MP4 → VideoCapture → Frame → Visualización

## 🔍 Explicación

- `VideoCapture("video.mp4")` → abre el archivo de video
- `cap.read()` → obtiene cada frame del video
- `imshow()` → muestra el frame
- `waitKey()` → controla velocidad y salida
- `ret=False` → indica fin del video

## ⚠️ Nota importante

- `cv2.imshow()` puede no funcionar bien en Jupyter
- Ejecutar en:
  - VS Code
  - Script Python (.py)

## 🚀 Aplicación en tu proyecto

Este mismo flujo se usa para:

Video → Gray → Blur → Canny → ROI → Hough → Lane Detection
```

#### 6 — OPCIONAL (PROCESAMIENTO EN TIEMPO REAL)

```python
while cap.isOpened():
    
    ret, frame = cap.read()
    
    if not ret:
        break
    
    # Convertir a escala de grises
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    # Blur
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    
    # Canny
    edges = cv2.Canny(blur, 50, 150)
    
    #resize window
    cv2.namedWindow("Edges", cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Edges", 800, 600)

    #resize window
    cv2.namedWindow("Test MP4", cv2.WINDOW_NORMAL)
    cv2.resizeWindow("Test MP4", 800, 600)

    # Mostrar
    cv2.imshow("Edges", edges)

    # Mostrar frame
    cv2.imshow("Test MP4", frame)
    
    if cv2.waitKey(1) == 13:
        break

cap.release()
cv2.destroyAllWindows()
```
