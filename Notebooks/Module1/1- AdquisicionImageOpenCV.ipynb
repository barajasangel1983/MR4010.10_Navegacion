#### Adquisición de Imagen y Espacios de Color (OpenCV)

#### 1. CARGA DE LIBRERIAS

```python
# ============================================
# 1. LIBRERÍAS
# ============================================
import cv2                      # OpenCV (visión computacional)
from PIL import Image           # Lectura alternativa de imágenes
import numpy as np              # Manejo de arreglos
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Mostrar imágenes dentro del notebook
%matplotlib inline
```

#### 2. CARGA DE IMAGEN Y DIMENSIONES

```python
# Leer imagen con OpenCV (formato BGR)
img = cv2.imread('D:\ML\Projects\Project_5_MR4010.10_Navegacion\MR4010.10_Navegacion\data\Module1\Self_Driving_Car.jpg')

# Validación
if img is None:
    raise ValueError("No se pudo cargar la imagen")

# Dimensiones
print("Shape:", img.shape)

print("Altura:", int(img.shape[0]), "pixeles")
print("Anchura:", int(img.shape[1]), "pixeles")
print("Canales:", int(img.shape[2]))
```

#### 3. VISUALIZACIÓN CON OPENCV

```python
# Muestra en ventana externa (no en notebook)
cv2.imshow('D:\ML\Projects\Project_5_MR4010.10_Navegacion\MR4010.10_Navegacion\data\Module1\Self_Driving_Car.jpg', img)
cv2.waitKey(0)          # Espera tecla
cv2.destroyAllWindows() # Cierra ventana
```

#### 4. VISUALIZACIÓN CON MATPLOTLIB

```python
# Convertir BGR → RGB (MUY IMPORTANTE)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Imagen RGB")
plt.axis("off")

# Ver dimensiones
print(img_rgb.shape)
```

🧠 Explicación:

OpenCV usa BGR

Matplotlib usa RGB

Si no conviertes → colores incorrectos

#### 5. CONVERSIÓN A ARRAY (NUMPY)

```python
# Convertir a arreglo
img_arr = np.asarray(img_rgb)

print("Shape array:", img_arr.shape)

plt.imshow(img_arr)
plt.title("Imagen como array")
plt.axis("off")
```

#### 6. LECTURA CON PIL (ALTERNATIVA)

```python
img_pil = Image.open('D:\ML\Projects\Project_5_MR4010.10_Navegacion\MR4010.10_Navegacion\data\Module1\Self_Driving_Car.jpg')

img_arr_pil = np.asarray(img_pil)

print(img_arr_pil.shape)

plt.imshow(img_arr_pil)
plt.title("Imagen con PIL")
plt.axis("off")
```

#### 7. EXTRACCIÓN DE CANALES (RGB)

```python
# Copia de la imagen
img_red = img_arr.copy()

# Eliminar canal verde y azul
img_red[:, :, 1] = 0
img_red[:, :, 2] = 0

plt.imshow(img_red)
plt.title("Canal Rojo")
plt.axis("off")
```

#### 8. CONVERSIÓN A HSV

```python
# Convertir BGR → HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Mostrar con OpenCV (opcional)
cv2.imshow('HSV', hsv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 9. CANAL HUE

```python
plt.imshow(hsv_img[:, :, 0], cmap='gray')
plt.title("Hue (Color)")
plt.axis("off")
```

🧠 Hue:

Representa el tipo de color

Ej: rojo, azul, verde

#### 10. CANAL SATURATION

```python
plt.imshow(hsv_img[:, :, 1], cmap='gray')
plt.title("Saturation")
plt.axis("off")
```

🧠 Saturation:

Intensidad del color

0 = gris, alto = color fuerte

#### 11. CANAL VALUE

```python
plt.imshow(hsv_img[:, :, 2], cmap='gray')
plt.title("Value (Brillo)")
plt.axis("off")
```

🧠 Value:

Brillo de la imagen

Muy importante en visión autónoma

#### RESUMEN CLAVE (IMPORTANTE)

BGR (OpenCV) ≠ RGB (Matplotlib)

HSV:

- H → Color

- S → Intensidad

- V → Brillo
